```python
import pandas as pd
import numpy as np
from pandas import read_csv
from matplotlib import pyplot
import matplotlib.pyplot as plt
#for AR
from statsmodels.tsa.ar_model import AutoReg
#for SES
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
#for HWES
from statsmodels.tsa.holtwinters import ExponentialSmoothing

%matplotlib inline
```

```python
#'[1h]BTCUSDT.csv'
data1 = pd.read_csv('./data/[1h]BTCUSDT.csv')
print('Origin Dataset: \n', data1.head())
```

```python
ax1 = data1['Close'].values
bx1 = pd.to_datetime(data1['Stop'].values, unit='ms').values
plt.figure(figsize=(10,10))
plt.plot(bx1,ax1)
```

# Autoregression

```python
model1 = AutoReg(ax1,lags=2).fit()
pred1 = model1.predict(0, len(ax1)+100)
plt.figure(figsize=(10,10))
plt.plot(range(len(pred1)), pred1, c='purple')
plt.plot(range(len(ax1)), ax1, c='yellow')
plt.show()
```

# Moving Average

```python
data1_close=pd.DataFrame(data1.Close)
data1_close['MA_21']=data1_close.Close.rolling(21).mean()
data1_close['MA_21'].head(21)
plt.figure(figsize=(10,10))
plt.grid(True)
plt.plot(data1_close['Close'], label='BTC')
plt.plot(data1_close['MA_21'], label='MA 21 day')
plt.legend(loc=2)
```

# Simple Exponential Smoothing

```python
mod1 = SimpleExpSmoothing(ax1).fit()
pred11 = mod1.predict(0, len(ax1)+100)
plt.figure(figsize=(15,10))
plt.plot(range(len(pred11)), pred11, c='blue')
plt.plot(range(len(ax1)), ax1, c='orange')
plt.show()
```

# Holt Winter’s Exponential Smoothing

```python
mode1 = ExponentialSmoothing(ax1).fit()
pred01 = mode1.predict(0, len(ax1)+100)
plt.figure(figsize=(15,10))
plt.plot(range(len(pred01)), pred01, c='blue')
plt.plot(range(len(ax1)), ax1, c='orange')
plt.show()
```
